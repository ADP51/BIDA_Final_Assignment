```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

df_yearly = pd.read_csv("./data_v2/yearly/2019.csv")
#drop unwanted/duplicate columns
df_yearly = df_yearly.drop(columns=['G', 'GS', 'Age', 'Yds', 'Yds.1', 'Yds.2', 'Int', 'Att', 'Att.1', 'Fumbles', 'FumblesLost', 'Cmp'])
#combine rushing attempts and passing attempts
df_yearly['Touches'] = df_yearly['Rec'] + df_yearly['RushingAtt']
df_yearly.head()
```

```python
# plot to see the correlation between the number of times a plyaer touches the ball to the amount of FP
plt.style.use('ggplot')
df_yearly.plot(x='Touches', y='FantasyPoints', kind='scatter')
```

```python
# we know there is a strong correlation between touches and points
# lets see which players were the most efficient with their touches
# create new skill position df
df_skillpos = df_yearly[(df_yearly.Pos == 'RB') | (df_yearly.Pos == 'WR') | (df_yearly.Pos == 'TE')] 
df_skillpos.head()
```

```python
df_skillpos['ppt'] = df_skillpos['FantasyPoints'] / df_skillpos['Touches']
df_skillpos.sort_values(by='ppt', ascending=False)
```

```python
# let's remove any player without at least 50 touches
df_skillpos.drop(df_skillpos[df_skillpos.Touches < 50].index, inplace=True)
df_skillpos.sort_values(by='ppt', ascending=False)
```

```python
# Let's do something similiar for QB's 
df_qbyearly = df_yearly[df_yearly.Pos == 'QB']
df_qbyearly.drop(columns='Touches')
df_qbyearly['Touches'] =  df_qbyearly['PassingAtt'] + df_qbyearly['RushingAtt']
df_qbyearly['ppt'] = df_qbyearly['FantasyPoints'] / df_qbyearly['Touches']
df_qbyearly.sort_values(by='ppt', ascending=False)
```

```python
# So now we have our ppt which shows us how efficient our players are with their attempts
# We still need to factor in the amount of touches with the efficiency to come up with a ranking system
# Let's say that for every 50 touches we give +1 to rank then times that by their ppt rating
df_skillpos['Rating'] = (df_skillpos['Touches']/50) * df_skillpos['ppt']
df_qbyearly['Rating'] = (df_qbyearly['Touches']/50) * df_qbyearly['ppt']
df_skillpos.sort_values(by='Rating', ascending=False)
```

```python
#split into training and test data 80/20
df_skilltrain = df_skillpos[:109]
df_skilltest = df_skillpos[-27:]

df_qbtrain = df_qbyearly[:57]
df_qbtest = df_qbyearly[-14:]
```

```python
#set up x and y for skillpos
skill_x_train = df_skilltrain.drop(columns=['Player','Tm', 'Pos'])
skill_x_test = df_skilltest.drop(columns=['Player', 'Tm', 'Pos'])

skill_y_test = df_skilltest['FantasyPoints']
skill_y_train = df_skilltrain['FantasyPoints']
skill_x_test.head()
```

```python
regr = linear_model.LinearRegression()
regr.fit(skill_x_train.values, skill_y_train.values)

skill_y_pred = regr.predict(skill_x_test)
```
